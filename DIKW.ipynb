### Correlation Analysis: Weather Data and Field B Yield Prediction

To understand how weather factors might influence crop yield, we'll calculate the correlation between key weather metrics and `field_B_yield_prediction_t`. Correlation coefficients range from -1 to 1:

*   **1**: Perfect positive linear relationship (as one variable increases, the other increases proportionally).
*   **-1**: Perfect negative linear relationship (as one variable increases, the other decreases proportionally).
*   **0**: No linear relationship.

```python
# 1. Load the daily dataset
df = pd.read_csv('complete-farm-stats-1998.csv')

# 2. Convert the 'date' column to proper datetime objects for plotting
df['date'] = pd.to_datetime(df['date'])

# 3. Sort the data by date just in case it's out of order
df = df.sort_values('date')

# 4. Calculate a 30-day rolling average to smooth out daily noise
df['rolling_yield'] = df['field_B_yield_prediction_t'].rolling(window=30).mean()

# 5. Set up the visualization
plt.figure(figsize=(14, 7))

# Plot the raw daily predictions (light blue, slightly transparent)
plt.plot(df['date'], df['field_B_yield_prediction_t'],
         label='Daily Prediction', color='#87CEEB', alpha=0.6, linewidth=1.5)

# Plot the 30-day smoothed trend line (dark blue, thicker line)
plt.plot(df['date'], df['rolling_yield'],
         label='30-Day Moving Average', color='#000080', linewidth=2.5)

# 6. Format the chart
plt.title('Field B Yield Prediction Over Time (1998)', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Date', fontsize=12, labelpad=10)
plt.ylabel('Yield Prediction (Tonnes)', fontsize=12, labelpad=10)

# Format the x-axis to show months clearly
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)

# Add gridlines and a legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left', fontsize=11)
plt.tight_layout()

# Add text annotations for key correlation findings
plt.text(0.02, 0.95, 'Key Correlations with Yield:', transform=plt.gca().transAxes, fontsize=10, verticalalignment='top')
plt.text(0.02, 0.90, '+ Rainfall: Moderate Positive (0.40)', transform=plt.gca().transAxes, fontsize=9, verticalalignment='top', color='green')
plt.text(0.02, 0.85, '- Daylight Hours: Moderate Negative (-0.48)', transform=plt.gca().transAxes, fontsize=9, verticalalignment='top', color='red')
plt.text(0.02, 0.80, '- Temperature: Weak Negative (~-0.32)', transform=plt.gca().transAxes, fontsize=9, verticalalignment='top', color='orange')

plt.show()
```

### Correlation Analysis: Weather Data and Field B Yield Prediction

To understand how weather factors might influence crop yield, we'll calculate the correlation between key weather metrics and `field_B_yield_prediction_t`. Correlation coefficients range from -1 to 1:

*   **1**: Perfect positive linear relationship (as one variable increases, the other increases proportionally).
*   **-1**: Perfect negative linear relationship (as one variable increases, the other decreases proportionally).
*   **0**: No linear relationship.

```python

# Identify weather-related columns and the yield prediction column
weather_cols = [
    'temperature_c',
    'rainfall_mm',
    'soil_moisture_pct',
    'soil_temp_field_A_c',
    'daylight_hours'
]
yield_col = 'field_B_yield_prediction_t'

# Combine into a single DataFrame for correlation calculation
correlation_df = df[weather_cols + [yield_col]]

# Calculate the correlation matrix
correlation_matrix = correlation_df.corr()

# Display correlations with yield prediction
yield_correlations = correlation_matrix[yield_col].drop(yield_col)
print("Correlation with Field B Yield Prediction:")
display(yield_correlations)
```

### Scatter Plots with Regression Lines: Environmental Factors vs. Field B Yield Prediction

These plots visualize the relationship between each environmental factor and the predicted yield. The regression line helps to highlight the linear trend, if any, and visually confirm the correlations we previously calculated.
